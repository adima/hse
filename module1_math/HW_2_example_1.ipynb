### Расстояние от точки до гиперплоскости
Расстояние от точки $P=(p_1, p_2, \ldots, p_n)\in \mathbb R^n$ до гиперплоскости $H$, заданной уравнением $a_1 x_1 + a_2 x_2 + \ldots + a_n x_n + b = 0$, можно вычислить по формуле:

$$d(P, H) = \frac{|a_1 p_1 + a_2 p_2 + \ldots + a_n p_n + b|}{\sqrt{a_1^2+a_2^2 + \ldots + a_n^2}}$$

### Задача 5
Рассмотрим $n$-мерное пространство. Найти расстояние от начала координат до гиперплоскости, проходящей через точки $(1, 0, \ldots, 0, 0)$, $(0, 1, \ldots, 0, 0)$,..., $(0, 0, \ldots, 1, 0)$, $(0, 0, \ldots, 0, 1)$ (единица на одном месте, остальные компоненты нулевые).

**Подсказка.** Сначала нужно найти уравнение гиперплоскости. Пусть искомое уравнение имеет вид $a_1 x_1 + a_2 x_2 + \ldots a_n x_n + b = 0$. Подставляя каждую из точек, получим систему уравнений на $a_1,a_2,\ldots, a_n, b$. Достаточно найти какое-нибудь решение этой системы (например, можно положить $b=1$ и найти все $a_i$, $i=1,\ldots, n$).

*Найдём уравнение гиперплоскости в общем виде: $a_1 x_1 + a_2 x_2 + \ldots a_n x_n + b = 0$.*

*Получим систему для всех i: $a_i + b = 0$.*

*Отсюда получаем искомое уравнение гиперплоскости:*

$-bx_1-bx_2+…-bx_n+b=0$

*Вынесем -b:*

$x_1+x_2+…x_n-1=0$

*Подставим всё в формулу расстояния от точки до гиперплоскости и получим ответ:*

$$d(P, H) = \frac{1}{\sqrt{n}}$$

### Задача 6

Докажите, что $(AB)^T=B^T A^T$ для любых матриц $A$, $B$, таких, что определено произведение $AB$. Здесь $X^T$ — транспонированная матрица $X$.

*На i-строке, j-столбце произведения матриц AB находится элемент вида: $\sum \limits_{k=1}^{n}a_{ik} b_{kj}$ , где n-число столбцов в A и строк в B*

*В транспонированной матрице этот элемент будет находится на j-строке, i-столбце.*

*Теперь запишем, что будет находится в матрице $B^T A^T$ на пересечении j-строки, i-столбца:* $\sum \limits_{k=1}^{n}b^t_{jk}a^t_{ki}$

*Т.к. $b^t_{jk}$ элемент транспонированной матрицы равен $b_{kj}$ исходной, получим следующее выражение для ji-го элемента в матрице $B^T A^T$:*

$B^T A^T (ji) = \sum \limits_{k=1}^{n}b_{kj}a_{ik} = \sum \limits_{k=1}^{n}a_{ik}b_{kj} = (AB)^T (ji)$

*Следовательно: $(AB)^T=B^T A^T$ для любых матриц $A$, $B$, таких, что определено произведение $AB$.*
