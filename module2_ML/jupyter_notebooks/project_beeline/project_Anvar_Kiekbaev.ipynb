###  Часть 2. Первичный анализ признаков

Общая информация по датасету:

```python
df.info()
```

Видим, что возможно есть количественные(float64) признаки, номинальные(object) признаки и количественные/номинальные/бинарные(int64) признаки.
Рассмотрим более подробно каждые признаки.

Общая статистика по всем количественным признакам:

```python
df.describe().T
```

Данные содержат **пропуски**.

Посмотрим статистику по признакам **x6**, **x7** и **x34**- так как min и max у них равен соответственно **0** и **1**, а тип - целочисленный, то скорее всего это бинарные признаки:

```python
print('Признак "x6":\n{}\n'.format(df['x6'].value_counts()))
print('Признак "x7":\n{}\n'.format(df['x7'].value_counts()))
print('Признак "x34":\n{}'.format(df['x34'].value_counts()))
```

Судя по статистике, данные **бинарные** и содержат **пропуски**.

Посмотрим статистику по номинальным признакам:

```python
nominal_stats = ['count', 'nunique', 'most_used_val', 'most_used_val_count']
df_nominal = pd.DataFrame([], [], columns=nominal_stats)
for col, typ in df.dtypes.iteritems():
    if typ == np.object:
        nominal_feature = df[col]
        df_nominal.loc[col] = [
            nominal_feature.count(),
            nominal_feature.nunique(),
            nominal_feature.value_counts().head(1).index[0],
            nominal_feature.value_counts().head(1).values[0]]
        
df_nominal
```

Данные содержат **пропуски**.


Так же имеются признаки, которые могут быть бинарными (**nunique == 2**)

```python
for col in df_nominal[df_nominal['nunique'] == 2].T.columns:
    print('Признак "{}":\n{}\n'.format(col, df[col].value_counts()))    
```

Имеем **41** количественный признак, **6** бинарных признаков с целочисленными и текстовыми значениями (**x2, x3, x6, x7, x15**), а так же **16** номинальных признаков c текстовыми значениями. Целевой признак y - номинальный (возможно порядковый) с целочисленными значениями.


```python
b_features = ['x2', 'x3', 'x6', 'x7', 'x15', 'x34']
n_features = [feature for feature in df_nominal.index if feature not in b_features]
q_features = [
    feature for feature in df.columns if feature not in b_features \
              and feature not in n_features \
              and feature != 'y'
]
```

Почти все признаки имеют сильные выбросы, нужно чистить. Так же многие признаки имеют одно или несколько значений, которое появляется на порядок чаще всех других. Попробуем эти значения тоже убрать. 

```python
for feature in q_features:
    df[feature + '_norm'] = df[feature]

df['x8_norm'] = df['x8'].apply(lambda x: None if x < -2 or x > 2 else x)
df['x13_norm'] = df['x13'].apply(lambda x: None if x < 0 or x > 15000 else x)
df['x23_norm'] = df['x23'].apply(lambda x: None if x < 10 or x > 60 else x)
df['x24_norm'] = df['x24'].apply(lambda x: None if x < -1 or x > 5 else x)
df['x25_norm'] = df['x25'].apply(lambda x: None if x < -5 or x > 5 else x)
df['x26_norm'] = df['x26'].apply(lambda x: None if x > 20 else x)
df['x27_norm'] = df['x27'].apply(lambda x: None if x < -5 or x > 5 else x)
df['x28_norm'] = df['x28'].apply(lambda x: None if x < -5 or x > 5 else x)
df['x29_norm'] = df['x29'].apply(lambda x: None if x > 100 else x)
# Убираем преобладающий 0
df['x30_norm'] = df['x30'].apply(lambda x: None if x <= 0 or x > 0.02 else x)
# Убираем преобладающий 0
df['x31_norm'] = df['x31'].apply(lambda x: None if x <= 0 or x > 300 else x)
df['x32_norm'] = df['x32'].apply(lambda x: None if x > 3 else x)
df['x33_norm'] = df['x33'].apply(lambda x: None if x > 60 else x)
df['x36_norm'] = df['x36'].apply(lambda x: None if x > 20 else x)
df['x37_norm'] = df['x37'].apply(lambda x: None if x > 400 else x)
df['x38_norm'] = df['x38'].apply(lambda x: None if x > 20 else x)
df['x39_norm'] = df['x39'].apply(lambda x: None if x > 20 else x)
df['x40_norm'] = df['x40'].apply(lambda x: None if x > 200 else x)
df['x43_norm'] = df['x43'].apply(lambda x: None if x > 10 else x)
df['x44_norm'] = df['x44'].apply(lambda x: None if x > 10 else x)
# Для x45 - x52 убираем преобладающие 0 и 1
for i in range(45, 53):
    df['x' + str(i) + '_norm'] = df['x' + str(i)].apply(lambda x: None if x == 1 or x == 0 else x)
# Убираем преобладающий -1.6
df['x53_norm'] = df['x53'].apply(lambda x: None if x < -1.6 else x)
# Убираем преобладающий -1.8
df['x54_norm'] = df['x54_norm'].apply(lambda x: None if x < -1.83865 else x)
# Для x45 - x52 убираем преобладающий 0
for i in range(55, 61):
    df['x' + str(i) + '_norm'] = df['x' + str(i)].apply(lambda x: None if x == 0 else x)
df['x61_norm'] = df['x61'].apply(lambda x: None if x > 1000 else x)
```

Все нормализованные количественные признаки:

```python
q_features_norm = [feature + '_norm' for feature in q_features]
```

```python
df[q_features_norm[:10]].hist(figsize=(14,10), bins=80);
```

```python
df[q_features_norm[10:20]].hist(figsize=(14,10), bins=80);
```

```python
df[q_features_norm[20:30]].hist(figsize=(14,10), bins=80);
```

```python
df[q_features_norm[30:]].hist(figsize=(14,10), bins=80);
```

Можно заметить, что признаки с x55 по x60 имеют одинаковый порядок значений [0, 1], а гистограммы их похожи. Посмотрим статистику и график их сумм.

```python
df['x55_60_sum'] = \
    df['x55_norm'] + df['x56_norm']  + df['x57_norm'] + df['x58_norm'] + df['x59_norm'] + df['x59_norm']

plt.plot(df[pd.notnull(df['x55_60_sum'])]['x55_60_sum'])
plt.ylabel('Sum of features x55, ..., x60')
plt.xlabel('Index')
plt.show();

print(df['x55_60_sum'].value_counts().head())
df['x55_60_sum'].hist(bins=100)
```

Сумма стремится к 1, то есть эти признаки в скорее всего что-то вроде долей целого.

Посмотрим корреляцию Пирсона между количественными признаками:

```python
corr = df[q_features_norm].corr().abs()

fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr, ax=ax)
```

Достаточно много признаков c существенным коэфицентом корреляции. Попробуем найти те признаки, где коэффициент корреляции выше 0.75. 

```python
corr = corr[corr > 0.75]
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr, ax=ax)
```

Пары признаков с максимальными коэффициентами корреляции:

```python
corr = corr[corr != 1]
# Воспроизводимость: если уже сделан unstack, то не нужно'''
if isinstance(corr, pd.core.frame.DataFrame):
    corr = corr.unstack()
corr = corr[pd.notnull(corr)]
corr = corr.order(ascending=False)
# Вывод без повторений
for c, k in corr[::2].iteritems():
    print(c, k)
```
